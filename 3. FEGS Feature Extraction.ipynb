## Dependencies

```python
import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings("ignore")
```

Features were extracted with the use "Feature Extraction based on Graphical and Statistical features" (FEGS). 
https://bmcbioinformatics.biomedcentral.com/articles/10.1186/s12859-021-04223-3

## Loading Data

```python
# loading the FEGS features
features = pd.read_csv('Data/features.csv', header=None)
features
```

```python
# adding ID to features
features.insert(loc=0, column='ID', value=range(1, len(features)+1))
features
```

## Merging features with original dataset

```python
dataset = pd.read_csv('Data/train_clean.csv')
dataset
```

```python
merged_data = dataset.merge(features, on='ID')
merged_data
```

```python
# changing the column names of the dataframe
column_names = list(merged_data.columns)
for i in range(4, len(column_names)): 
    column_names[i] = 'f' + str(i-3)
merged_data = merged_data.set_axis(column_names, axis=1, inplace=False)
merged_data
```

```python
# converting dataframe to CSV
merged_data.to_csv('Data/main_dataset.csv', index=False)
```
