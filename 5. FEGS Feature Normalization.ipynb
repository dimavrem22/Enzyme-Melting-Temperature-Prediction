## Dependencies

```python
import numpy as np
import pandas as pd
```

## Loading Data

```python
training_data = pd.read_csv('Data/training_data.csv')
training_data
```

```python
testing_data = pd.read_csv('Data/testing_data.csv')
testing_data
```

## Normalizing Training Dataset

```python
from sklearn import preprocessing

# converting features to numpy array
features_list = list(training_data.columns)[4:]
feature_training_array = training_data[features_list].to_numpy()

# fitting the data into Gaussian distribution
scaler = preprocessing.StandardScaler().fit(feature_training_array)

# transforming features features of training data
training_data[features_list] = scaler.transform(feature_training_array)

# saving normalized data
training_data.to_csv('Data/traning_data_normalized.csv', index=False)

training_data
```

## Normalizing Testing Data

```python
# using training scalar to transform testing data
feature_test_array = testing_data[features_list].to_numpy()

# transforming features features of training data
testing_data[features_list] = scaler.transform(feature_test_array)

# saving normalized data
testing_data.to_csv('Data/testing_data_normalized.csv', index=False)

testing_data
```
